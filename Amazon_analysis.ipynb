This might or might not give us useful insight into the exact sales numbers we are looking for so we can instead try to 
plot a scatter plot for all the categories to check if it is more insightful.

```python
# Count the frequency of each category
category_counts = {}
for category in data['category']:
    for category in category.split('|'):
        if category in category_counts:
            category_counts[category] += 1
        else:
            category_counts[category] = 1

categories = []
counts = []
for category, count in sorted(category_counts.items(), key=lambda x: x[1], reverse=False):
    categories.append(category)
    counts.append(count)
```

```python
colors = []
for count in counts:
    if count < 50 and count:
        colors.append('red')
    elif count >= 50 and count <= 200:
        colors.append('yellow')
    else:
        colors.append('green')
        
plt.scatter(categories, counts, c=colors)
plt.title('How many items are popular at amazon?')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.xticks([])
plt.legend()
```

From the graph we can infer that there are not a lot of categories that sell very well on amazon , red represents the sale count 
of less than 50 , yellow is 50 to 200 and green is everything over that , to see the top categories we can probably make a wordcloud 
to be visually pleasing.

```python
from wordcloud import WordCloud

category_counts = {}


for category in data['category']:
    for category in category.split('|'):
        if category in category_counts:
            category_counts[category] += 1
        else:
            category_counts[category] = 1

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(category_counts)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show
```

```python
data[['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5', 'Category 6', 'Category 7']] = data['category'].str.split('|', expand=True)

#This was done because when describing a product the category can go like following - computers | Accessories | Cables and so on
#maximum i saw was 6 sub categories to a category but i made 7 just in case there are more sub categories , these can always be changed later
#down the road.

data.drop('category', axis=1, inplace=True)
```

```python
data.head()
```

```python
data.loc[data["rating"].isin(["|"])] = "0"

#changing all the columns that need to be changed 

columns_need_change = [ "discounted_price"  ,  "actual_price"   ,  "discount_percentage"   ,  "rating"  ,   "rating_count" ]

for i in columns_need_change:
    data[i] = data[i].str.replace("₹","")
    data[i] = data[i].str.replace("%","")
    data[i] = data[i].str.replace(",","")

    data[i] = data[i].astype("float")
    
    
```

```python
data.drop_duplicates(subset=["product_id"],inplace=True)
```
